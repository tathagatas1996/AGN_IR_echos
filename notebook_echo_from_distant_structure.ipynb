```python
import numpy as np
from dustecho import *
import matplotlib.pyplot as plt
```

```python
Tmin = 0
Tmax = 5000
t  = np.arange( Tmin , Tmax, 10.0)*days_to_sec
```

## Testing for different values of inner radii

```python
"""
1. Radii: Rin = 0.2 parsec, Rout = 1.0 parsec, opening angle = 10 degree
"""
lc10 = dustecho("biconically-cutout-shell", 1)
lc10.thetao = 10.*np.pi/180
lc10.rin = 0.2*pc
lc10.rout= 1.0*pc
lc10.set_the_input_lc_params(1)
flux10 = lc10.f0(t)


"""
2. Radii: Rin = 0.4 parsec, Rout = 1.0 parsec, opening angle = 10 degree
"""
lc20 = dustecho("biconically-cutout-shell", 1)
lc20.thetao = 10.*np.pi/180
lc20.rin = 0.4*pc
lc20.rout= 1.0*pc
lc20.set_the_input_lc_params(1)
flux20 = lc20.f0(t)
```

```python
plt.figure(figsize=(8,6))
plt.plot( t/days_to_sec , flux10/min(flux10), label=r"$R_{in}$ = 0.2 pc", color='blue')
plt.plot( t/days_to_sec , flux20/min(flux20), label=r"$R_{in}$ = 0.4 pc", color='magenta')
plt.plot( t/days_to_sec , lc10.Lin(t)/min(lc10.Lin(t)), label=r"Input X-ray", color='k')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel("Time [days]", fontsize=15)
plt.ylabel("Flux [arbitrary units]", fontsize=16)
plt.legend(loc='best', fontsize=15)
plt.xlim(0,3000)
```
